Exercise 1

```python
name = input("Enter your name: ")
age = int(input("Enter your age: "))
height = int(input("Enter your height in cm: "))
weight = int(input("Enter your weight in kg: "))

# Print lowercase, uppercase, and length of the name
lower_name = name.lower()
print(f"\n\nLowercase name: {lower_name}")
upper_name = name.upper()
print(f"Uppercase name: {upper_name}")
name_length = len(name)
print(f"Length of name: {name_length}")

# Convert height to meters and print BMI
height_in_meters = height / 100
print(f"BMI: {weight / (height_in_meters ** 2)}")

# Print weight divided by pi
import math
print(f"Weight divided by pi: {weight / math.pi}")

# Print age in days
age_in_days = age * 365
print(f"Approximate age in days: {age_in_days}")

# Split name into characters and print them one per line
character_list = list(name)
print("Name split into characters:")
for item in character_list:
    print(f"   {item}")
```

```python
num1 = int(input("Enter a number: "))
num2 = int(input("Enter another number: "))

# Sum
sum = num1 + num2
print(f"Sum: {sum}")

# Difference
diff = num1 - num2
print(f"Difference: {diff}")
      
# Product
product = num1 * num2
print(f"Product: {product}")
      
# Quotient (division result)
quotient = num1 / num2
print(f"Quotient: {quotient}")
```
